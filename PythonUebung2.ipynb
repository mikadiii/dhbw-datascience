# Warum Pandas?
- Übung basierend auf Material von Prof. Dietzsch
- Pandas ist eine der leistungsfähigsten Bibliotheken, die man für Data Science in Python braucht<br>
- Pandas bietet eine schnelle, flexible, leichte und intuitive Art der Manipulation von strukturierten/tabellenförmigen Daten (Data-Frames)<br> 
- Es ermöglicht die Arbeit mit großen Datensätzen und/oder mehreren Dateien zugleich.<br>


## Was zeichnet Pandas aus?

- `pandas` baut auf `numpy` auf und nutzt dieses als Backend
- Stellt eine High-Level Daten-Struktur zur Verfügung, die Spreadsheets von Tabellenkalkulationen ähnelt
- Verfügt über eingebaute Funktionen, um Daten zu bereinigen, zu filtern, zu gruppieren, zu kombinieren ... etc. 
- Es bietet die einfache Möglichkeit `numpy`- und `scipy`-Funktionen auf Datentabellen anzuwenden

```python
#Aufgabe 5
"""
print(pokemon_data[['Type 1','Type 2']].value_counts().head(5),'\n')
print('Type 1:\n',pokemon_data['Type 1'].value_counts().head(5),'\n')
print('Type 2:\n',pokemon_data['Type 2'].value_counts().head(5),'\n')"""
ax = plt.axes()
ax.set_title('Heatmap of Type Pairs')
_=sns.heatmap(pokemon_data.groupby(['Type 1','Type 2']).size().unstack(),ax=ax)
```

```python
#Aufgabe6 -- kein Plan wie Pokemon Attaken funktionieren
_=pokemon_data[['Speed','Attack']].plot(kind='scatter',x='Speed',y='Attack',title='Scatterplot of Attack/Speed')
```

```python
# Aufgabe 7
atk_rating = pokemon_data['Attack'] + pokemon_data['Sp. Atk'] + pokemon_data['Speed']
def_rating = pokemon_data['Defense'] + pokemon_data['Sp. Def'] + pokemon_data['HP']
pokemon_data['atk_rating'] = atk_rating
pokemon_data['def_rating'] = def_rating

agro = pokemon_data['atk_rating'] - pokemon_data['def_rating']
pokemon_data['agro'] = agro
```

```python
# Offensive Pokemons
pokemon_data.sort_values(by=['agro'],ascending=False).head(10)
```

```python
# Defensive Pokemons
pokemon_data.sort_values(by=['agro'],ascending=True).head(10)
```

```python
#Aufgabe 8

_=pokemon_data.groupby('Generation').mean().iloc[:,1:7].plot(title='Stats per Generation',ylabel='points')
```

```python
# Aufgabe8 - Types
ax=plt.axes()
ax.set_title('HeatMap of Primary Type over Generations')
_=sns.heatmap(pokemon_data.groupby(['Type 1','Generation']).size().unstack(), ax = ax)
```

```python
# Aufgabe 8 - Overall
_=pokemon_data.groupby('Generation').mean().iloc[:,8].plot(xlabel='Generation',ylabel='mean(General)',title='General Score over the Generations')
```
